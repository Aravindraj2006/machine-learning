# CSV

```python
from io import StringIO,BytesIO
```

```python
data = ('col1,col2,col3\n'
'x,y,1\n'
'a,b,2\n'
'c,d,3')
```

```python
type(data)
```

```python
pd.read_csv(StringIO(data))
# create a text memory buffer if you give inside StringIO and to read it as csv file
```

### to read the specific column

```python
# read from specific column
df = pd.read_csv(StringIO(data),usecols = ['col1','col2'])
```

```python
df
```

## to convert data into csv file

```python
df.to_csv('test.csv')
```

```python
pd.read_csv('test.csv')
```

```python
data1 = ('a,b,c,d\n'
         '1,2,3,4\n'
         '5,6,7,8\n'
         '9,10,11,12\n')
```

```python
df = pd.read_csv(StringIO(data1),dtype = object)
# object means string each element are string
```

```python
df
```

```python
df['a'][0]
```

## to convert each column into specific data type

```python
df = pd.read_csv(StringIO(data1),dtype = {'a':int,'b':object,'c':float})
```

```python
df
```

```python
df['d']
```

```python
df.dtypes
# to check the data type
```

## to give index to the datatype

```python
data2 = ('z,a,b,c\n'
         '2,macaw,eagle,4\n'
         '3,cheetah,lion,0.7\n')
```

```python
df = pd.read_csv(StringIO(data2),index_col = 'z')
```

```python
df
```

```python
data3 = ('a,b,c\n'
         '4,apple,orange,\n'
         '8,bat,man,\n')
```

```python
df
# if you type comma on the last of your data the first element will go to the index
```
