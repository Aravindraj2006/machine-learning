```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("tips")
```

```python
data.head()
```

```python
# tip is dependent and bill,sex,day,time are independent
```

# Correlation with heatmap
### correlation is to find relationship 

```python
# only value are integer and float
```

```python
df = pd.DataFrame(data)

# Select only numeric columns
df_numeric = df.select_dtypes(include=['number'])

# Compute the correlation matrix
ar = df_numeric.corr()
```

```python
ar
```

```python
sns.heatmap(ar)
```

# jointplot
#### jointplot allows to study the relationship betweeb 2 numeric variables.the centre char display the correlation

## univarite analysis

```python
sns.jointplot(x = 'tip',y = 'total_bill',data = data,kind = 'hex') 
```

```python
# From the diagram we understand that the majority of them give tip as 2 and major total bill is between 10 to 20
```

```python
sns.jointplot(x = 'tip',y = 'total_bill',data = data,kind = 'reg')
```

```python
# we can see the best fit line
```
